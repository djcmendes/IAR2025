```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Results

## Evolutionary Braitenberg Machine

```python
# Carrega o CSV gerado pelo script
df = pd.read_csv('./thymio/controllers/p1_phase1_controller_genetic/fitness_history_Braitenberg.csv')

# Cria o grfico da mdia de fitness
plt.figure(figsize=(8, 6))
plt.plot(df['Generation'], df['AverageFitness'], label='Average Fitness')  # sem marker
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Fitness Evolution over Generations')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
